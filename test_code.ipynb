```python
import spacy

# 加载spaCy的语言模型
nlp = spacy.load("zh_core_web_sm")  # 如果是中文，使用zh_core_web_sm或其他支持的中文模型

def classify_question_nlp(question):
    # 定义问句分类关键词
    categories = {
        "地点": ["哪里", "哪儿", "地方", "地点"],
        "时间": ["何时", "时候", "时间", "日期", "几点"],
        "人物": ["谁", "名字", "姓名", "人物"],
        "数量": ["多少", "几", "数量", "多大"],
        "方法": ["如何", "怎样", "怎么", "方式", "方法"],
        "判断": ["是否", "吗", "能否", "会不会", "可以", "是"],
        "选择": ["哪个", "哪种"]
    }

    # 使用spaCy解析句子
    doc = nlp(question)

    # 遍历定义的分类
    for category, words in categories.items():
        # 遍历问句中的单词，与分类关键词对比
        if any(word in question for word in words):
            return category

        # 检查问句的主句结构，例如疑问词
        for token in doc:
            if token.pos_ == "PRON" or token.pos_ == "ADV":  # 处理疑问代词或副词
                if token.text in words:
                    return category

    # 如果无法分类
    return "无法分类"

# 示例问句
questions = [
    "他何时去上班？",
    "这个地方在哪里？",
    "这件事谁负责？",
    "总共有多少人参加？",
    "如何解决这个问题？",
    "这件事是否可以完成？"
]

# 分类输出
for q in questions:
    print(f"问句: {q} -> 类型: {classify_question_nlp(q)}")
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import torch

# 加载预训练的BERT模型和分词器
tokenizer = BertTokenizer.from_pretrained('bert-base-chinese')
model = BertForSequenceClassification.from_pretrained('bert-base-chinese')
```

```python
# 准备文本
text1 = "我喜欢你"
text2 = "我不喜欢你"

# 对每个文本进行编码
encoded_input1 = tokenizer(text1, return_tensors='pt')
encoded_input2 = tokenizer(text2, return_tensors='pt')

# 通过模型获取文本的向量表示
with torch.no_grad():
    output1 = model(**encoded_input1)
    output2 = model(**encoded_input2)

# # 取每个输出的最后一层的平均向量
sentence_embedding1 = output1.last_hidden_state.mean(dim=1).squeeze()
sentence_embedding2 = output2.last_hidden_state.mean(dim=1).squeeze()

print(len(sentence_embedding1))
# print(sentence_embedding2)

# # 计算两个向量之间的余弦相似度
similarity = 1 - cosine(sentence_embedding1.numpy(), sentence_embedding2.numpy())

print(f"相似度: {similarity}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

tokenizer = AutoTokenizer.from_pretrained('cross-encoder/ms-marco-TinyBERT-L-2')
model = AutoModelForSequenceClassification.from_pretrained('cross-encoder/ms-marco-TinyBERT-L-2')

user_question = "非洲野驴和印度野驴有一些区别，其中印度野驴栖息在哪儿？"
candidates = [
    "印度野驴栖息在哪儿？",
    "印度野驴与非洲野驴的分别是什么？"
]
inputs = tokenizer([f"{user_question} [SEP] {c}" for c in candidates], return_tensors='pt', padding=True, truncation=True)
scores = model(**inputs).logits
best_match_idx = torch.argmax(scores)
best_match = candidates[best_match_idx]
```
